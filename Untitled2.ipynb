```python
def palindrome(numb):
    sum = 0
    a = numb
    while a!=0:
        rem = a%10
        a = a//10
        sum = sum*10+rem
    if sum==numb:
        return "Palindrome"
    else:
        return "not Palindrome"


def frs_plndrm(num):
    lis1 = []
    dt1 = 10
    while len(lis1)<=num:
        if palindrome(dt1) == "Palindrome":
            lis1.append(dt1)
        dt1+=1
    return lis1
    
inp = int(input())
frs_plndrm(inp)


    
        
```

```python
def nest_list(n1, li1 = [], cnt=1):
    if n1==0:
        return li1
    if  cnt==1:
        li1 = [n1]*n1
    else:
        temp = [n1]*n1
        temp.append(li1)
        li1 = temp
    return nest_list(n1-1, li1, cnt+1)    
        
```
